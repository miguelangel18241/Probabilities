```python
import numpy as np

class WeightedNN:
    def __init__(self):
        pass

    def fit(self, X_train, y_train):
        self.X_train = X_train
        self.y_train = y_train

    def euclidean_distance(self, a, b):
        return np.linalg.norm(a - b)

    def weighted_vote(self, X_test):
        weighted_votes = np.zeros((X_test.shape[0],))
        for i, x_test in enumerate(X_test):
            distances = np.array([self.euclidean_distance(x_test, x_train) for x_train in self.X_train])
            weights = 1 / distances
            weighted_votes[i] = np.sum(weights * self.y_train) / np.sum(weights)
        return weighted_votes.astype(int)
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the range of distances
distances = np.linspace(0, 10, 100)

# Define the parameter controlling the rate of decay
power_parameter = 2

# Calculate the weights using the alternative weighting function
weights = 1 / distances**power_parameter

# Plot the weights
plt.plot(distances, weights)
plt.xlabel('Distance')
plt.ylabel('Weight')
plt.title('Weight Decay Curve with Alternative Function')
plt.show()
```
