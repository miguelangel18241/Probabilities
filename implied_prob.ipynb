```python
import random
import math
import pandas as pd
```

```python
# implied probability = 1/the probability
# sum the probabilities
# Normalize the prob by dividing the implied prob by the sum of the probabilities
```

```python
odds = {
    'Manchester United' : 2.10,
    'Liverpool' : 3.25,
    'Arsenal' : 3.50
}

# Implied

total_prob = sum(1/odds[outcome] for outcome in odds)
normalize = {outcome: (1/odds[outcome])/total_prob for outcome in odds}

total_stake = 100

prob_profit = {outcome: (100/normalize[outcome])*odds[outcome] for outcome in odds}

```

```python
def calculate_bets(odds, total_stake):
  total_prob = sum(1/odds[outcome] for outcome in odds)
  normalize = {outcome : (1/odds[outcome])/total_prob for outcome in odds}
  bet_amounts = {outcome : normalize[outcome]*total_stake for outcome in odds}
  potential_profits = {outcome : bet_amounts[outcome] * odds[outcome]-total_stake for outcome in odds}

  #Creating a dictionary for each team with odds, bet amount and profit
  display_bets = {}
  for outcome in odds:
    display_bets[outcome] = {
        'Odds' : odds[outcome],
        'Bet_amount' : bet_amounts[outcome],
        'Profit' : potential_profits[outcome]
    }
  return display_bets

odds = {'Liverpool': 3, 'Arsenal': 3.50}
stake = 100

display_bets = calculate_bets(odds, stake)

df = pd.DataFrame.from_dict(display_bets, orient='index')
df
```

```python
## This code is going to calculate the Area and circunference of a circle
# Define function
# Make an empty list
# Iterate the following:
  # Calculate Area
  # Calculate Circun
  # Append them the properties passed as a dictionary with the R, A and C
  # Return properties
# Make the random R list
# Call the function and save it in a var
# Make it a dataframe
```

```python
pi_value = math.pi

def calculate_circle_properties(list_radius):
  properties = []
  for radius in list_radius:
    area = pi_value*radius**2
    circun = 2*pi_value*radius
    properties.append({'Radius': radius, 'Area': area, 'Circunference': circun})
  return properties

rad_list = [random.uniform(1,10) for _ in range(20)]

properties_circle = calculate_circle_properties(rad_list)

df_circle = pd.DataFrame(properties_circle)
df_circle
```
