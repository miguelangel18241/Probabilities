```python
import pandas as pd
df = pd.read_csv('dataset_1.csv')
df.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
from mpl_toolkits.mplot3d import Axes3D
from sklearn.cluster import KMeans

# Perform KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
clusters = kmeans.fit_predict(df)

# 3D ScatterPlot
fig = plt.figure(figsize=(20, 10))
ax = fig.add_subplot(111, projection='3d')

# Colors
colors = ['gold', 'skyblue', 'red']

# Plotting based on the cluster
for cluster_id in range(3):
    cluster_data = df[clusters == cluster_id]
    ax.scatter(cluster_data.iloc[:, 0], cluster_data.iloc[:, 1], cluster_data.iloc[:, 2], color=colors[cluster_id], label=f'Cluster # {cluster_id+1}')

# Labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('ScatterPlot with Clusters in 3D\n The colors of my Country Flag')

# Add legend
ax.legend()
plt.show()
```
