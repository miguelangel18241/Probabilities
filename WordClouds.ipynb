```python
# Import necessary libraries
import numpy as np
from wordcloud import WordCloud
import matplotlib.pyplot as plt

# Define the text data
text = """
Happy with my furniture purchases. Struggle to find specific items in the store. Often ask for assistance from store staff. Purchase small items at Kmart (kitchen, bathroom appliances). Build furniture and feel satisfied afterward. Navigate through the shopping experience smoothly. Satisfaction after successfully building furniture. Content with the shopping experience. Concerned about the lack of diversity in Ikea's designs. Unsure about the quality of Ikea's products. Appreciative of the storage options available. Confusion navigating through the store. Struggle with lifting heavy furniture. Desire for more diverse furniture designs at Ikea. Questioning the quality of Ikea's products. Satisfied with the storage options Ikea provides. Find the Ikea store layout confusing. Difficulty with lifting heavy furniture items.
"""

# Define the IKEA blue and yellow colors
ikea_blue = "#0051ba"
ikea_yellow = "#ffda21"

# Define a function to generate color gradients
def color_func(word, font_size, position, orientation, random_state=None, **kwargs):
    if np.random.rand() < 0.5:
        return ikea_blue
    else:
        return ikea_yellow

# Generate the word cloud with custom colors
wordcloud = WordCloud(width = 800, height = 800,
                      background_color = 'white',
                      color_func = color_func,
                      contour_color = 'black',
                      contour_width = 2,
                      stopwords = None,
                      min_font_size = 10).generate(text)

# Plot the word cloud
plt.figure(figsize = (8, 8), facecolor = None)
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
plt.tight_layout(pad = 0)

# Show the plot
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Define the words and their respective sizes
words = ['furniture', 'store', 'shopping', 'quality', 'diversity', 'storage', 'confusion', 'heavy', 'designs', 'satisfaction']
sizes = [100, 80, 70, 60, 50, 100, 80, 70, 60, 50,]  # Size of bubbles, you can adjust as needed

# Define random positions for bubbles
x = np.random.rand(len(words))
y = np.random.rand(len(words))

# Define the IKEA blue and yellow colors
ikea_blue = "#0051ba"
ikea_yellow = "#ffda21"

# Plot the bubbles
plt.figure(figsize=(10, 6))
for i in range(len(words)):
    color = ikea_blue if i % 2 == 0 else ikea_yellow  # Alternating between blue and yellow
    plt.scatter(x[i], y[i], s=sizes[i] * 10, color=color, alpha=0.5, edgecolors='black')

    # Adjust the text position to be inside the bubble
    plt.text(x[i], y[i], words[i], fontsize=12, ha='center', va='center', color=color)

# Customize plot appearance
plt.title('Word Bubbles')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.xticks([])
plt.yticks([])
plt.axis('off')
plt.tight_layout()

# Show the plot
plt.show()
```
